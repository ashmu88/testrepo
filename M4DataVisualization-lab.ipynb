## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt

# Load the data into a pandas DataFrame
df = pd.read_sql_query("SELECT * FROM master", conn)

# Ensure that the 'ConvertedComp' column is numeric
df['ConvertedComp'] = pd.to_numeric(df['ConvertedComp'], errors='coerce')

# Plotting the histogram for the 'ConvertedComp' column
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=50, color='skyblue', edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import seaborn as sns 

# Plotting the box plot for the 'Age' column
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Age'], color='lightgreen')

# Adding title and labels
plt.title('Box Plot of Age')
plt.xlabel('Age')

# Show the plot
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

# Plotting the scatter plot for 'Age' vs 'WorkWeekHrs'
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], color='purple', alpha=0.6)

# Adding title and labels
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')

# Show the plot
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
# Ensure that 'Age', 'WorkWeekHrs', and 'CodeRevHrs' are numeric
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['WorkWeekHrs'] = pd.to_numeric(df['WorkWeekHrs'], errors='coerce')
df['CodeRevHrs'] = pd.to_numeric(df['CodeRevHrs'], errors='coerce')

# Drop rows with missing values
df = df.dropna(subset=['Age', 'WorkWeekHrs', 'CodeRevHrs'])

# Bubble plot: Use Age as bubble size
plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*10, alpha=0.5, color='blue')

# Adding title and labels
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs (Size by Age)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')

# Show the plot
plt.show()
```

```python
# Load the relevant column from the 'DatabaseWorkedWith' table (adjust column name if needed)
# Load the relevant columns from the 'DatabaseWorkedWith' table
df = pd.read_sql_query("SELECT Respondent, DatabaseWorkedWith, count(DatabaseWorkedWith) FROM DatabaseWorkedWith group by Respondent having count(DatabaseWorkedWith) = 1", conn)

# Filter rows where 'DatabaseWorkedWith' contains only 'MySQL' (no other databases)
mysql_only = df['DatabaseWorkedWith'].apply(lambda x: len(x.split(',')) == 1 and 'MySQL' in x)

# Count how many respondents work only with 'MySQL'
mysql_only_count = mysql_only.sum()

# Display the result
print(f"Number of respondents working with MySQL only: {mysql_only_count}")
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

# Load the relevant columns from the 'master' table
df = pd.read_sql_query("SELECT Age, WorkWeekHrs, CodeRevHrs FROM master", conn)

# Filter the data for age group 30 to 35
df_filtered = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate the median of WorkWeekHrs and CodeRevHrs for this age group
median_values = df_filtered[['WorkWeekHrs', 'CodeRevHrs']].median()

# Plotting the stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(median_values.index, median_values['WorkWeekHrs'], label='WorkWeekHrs', color='lightblue')
plt.bar(median_values.index, median_values['CodeRevHrs'], bottom=median_values['WorkWeekHrs'], label='CodeRevHrs', color='lightgreen')

# Adding title and labels
plt.title('Stacked Bar Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.ylabel('Hours')
plt.xlabel('Metrics')

# Display the legend
plt.legend()

# Display the plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import matplotlib.pyplot as plt


# Load the relevant columns from the 'master' table
df = pd.read_sql_query("SELECT Age, ConvertedComp FROM master", conn)

# Filter the data for age group 45 to 60
df_filtered = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Group by age and calculate the median of ConvertedComp
age_median = df_filtered.groupby('Age')['ConvertedComp'].median()

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(age_median.index, age_median.values, marker='o', color='b', linestyle='-', linewidth=2, markersize=6)

# Adding title and labels
plt.title('Median ConvertedComp for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')

# Display the plot
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
# Load the relevant column from the 'master' table (adjust column name if needed)
df = pd.read_sql_query("SELECT MainBranch FROM master", conn)

# Count the occurrences of each value in 'MainBranch'
mainbranch_counts = df['MainBranch'].value_counts()

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='skyblue', edgecolor='black')

# Adding title and labels
plt.title('Distribution of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')

# Display the plot
plt.show()
```
